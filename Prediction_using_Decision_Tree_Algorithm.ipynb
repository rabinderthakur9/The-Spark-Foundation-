Problem defination: Createating the Decision Tree classifier on Iris dataset and visualizing it graphically.

```python
# Importing libraries in Python
from sklearn.metrics import confusion_matrix 
from sklearn.tree import DecisionTreeClassifier 
import sklearn.metrics as metrics
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import tree
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#Loading datasets 
iris_data = datasets.load_iris() 
iris=pd.DataFrame(iris_data.data)

#priting features name of iris data 
print ("Features Name : ", iris_data.feature_names) 

#shape of datasets 
print ("Dataset Shape: ", iris.shape) 

#first five sample 
print ("Dataset: ",iris.head()) 
```

```python
iris_data.keys()
```

```python
#Checking the target values
iris_data.target_names
```

```python
#Splittig the data into variable
X = iris.values[:, 0:4] 
Y = iris_data.target

print(X)
print(Y)
```

## Splitting the Model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

## Importing and training the Decision tree Model

```python
# Decision tree classifier 
clf= DecisionTreeClassifier(random_state = 10)

#fitting the training data
clf.fit(X_train, y_train)
```

## Predicting On Random Sample

```python
# prediction on random data
X=[[6.4,1.8 ,6.6 ,2.1]]

Y_pred=clf.predict(X)

print(Y_pred)
```

```python
#We can see the index value for random sample output here of y_pred
iris_data.target_names
```

```python
# prediction on X_test (testing data )

Y_pred=clf.predict(X_test)

print(Y_pred)
```

```python
#Accuray of the model 
print("Accuracy:",metrics.accuracy_score(y_test, Y_pred))

cm=np.array(confusion_matrix(y_test,Y_pred))

cm
```

```python
fig=plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```
